# Education Performance Analysis

## Background
Education outcomes are influenced by various social and academic factors.
This project analyzes how parents' education levels relate to students' final academic performance.

## Dataset
The dataset contains simulated student records including:
- Gender
- Parents' education level
- Study hours
- Final academic score

## Objective
To explore the relationship between parents' education levels and students' academic performance using basic data analysis techniques.

```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

data = """
gender,parent_education,study_hours,final_score
Male,High School,2,68
Female,Bachelor,4,82
Female,Master,5,90
Male,High School,1,60
Female,Bachelor,3,78
Male,Associate,3,75
Female,Associate,4,80
Male,Master,5,88
"""

df = pd.read_csv(StringIO(data))
df

df.groupby("parent_education")["final_score"].mean()

df.groupby("parent_education")["final_score"].mean().plot(kind="bar")
plt.title("Average Final Score by Parents' Education Level")
plt.xlabel("Parents' Education")
plt.ylabel("Average Final Score")
plt.show()
```

## Key Insights

- Students whose parents have higher education levels tend to achieve higher final scores.
- This suggests that educational environment and parental academic background play an important role in student performance.
- Such insights highlight the importance of equal access to educational support for students from different backgrounds.
